### product pricing

```python
a = int(input("base demand = "))
b = int(input("price sensitivity = "))
c = int(input("unit cost = "))

max_profit = 0
optimal_price = 0

# p = price; c = cost
for p in range(c + 1, a // b):
    profit = (a - b * p) * (p - c)
    # print(p, profit)

    if profit > max_profit:
        max_profit = profit
        optimal_price = p
    else:
        break
    
print("optimal price = " + str(optimal_price))
print("maximized profit = " + str(max_profit))
```

```python
import math

bad = 0
for i in range(100):
    # pow(x, y): 計算 x^y

    f = pow(i, 1/2)
    if abs(f * f - i) > 0.000000000000001:
        print("!!!")
        bad += 1
    else:
        print(i)
print("bad precision:", bad)
```
