```python
def over_mother():
    print("Over the hills and far away")
    print("Mother duck said quack quack quack")
def n_ducks(num):
    print("%s little ducks went out one day" % num)
def only_n(num):
    print("But only %s little ducks came back" % num)
    
n_ducks("Five")
over_mother()
only_n("Four")
print()
n_ducks("Four")
over_mother()
only_n("Three")
print()
n_ducks("Three")
over_mother()
only_n("Two")
```

```python
def print_msg(item, ndays=30):
    # %s: string; %d: decimal integer
    print("Please return %s in %d days" % (item, ndays))
    print(f"Please return {item} in {ndays} days")

print_msg("items")
print_msg("items", 50)
print_msg(ndays=44, item = "bag")
```

### Recursion

```python
def factorial(n):
    if not isinstance(n, int):
        print("Factorial is only defined for integers")
        return None
    elif n < 0:
        print("Factorial is not defined for negative integers")
        return None
    elif n == 0: 
        return 1
    else:
        return n * factorial(n-1)
factorial(1.5)
```
