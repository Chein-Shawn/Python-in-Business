### Dictionary

```python
eng2cn = dict()
print(eng2cn)

eng2cn['one'] = '一'
eng2cn['two'] = '二'
eng2cn['three'] = '三'
eng2cn['four'] = '四'
print(eng2cn)
print(eng2cn['two'])
print(eng2cn.get('five', 'Not Found'))
print(eng2cn.get('five'))
print(eng2cn.setdefault('four'))
print(eng2cn.setdefault('five', '五'))
print(eng2cn)
print(eng2cn.keys())
print(eng2cn.values())
```

```python
def histogram(seq):
    d = dict()
    for i in seq:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    return d
print(histogram('parrot'))
for key in histogram('parrot'):
    print(key, histogram('parrot')[key])
print("-------------------")
for key, value in histogram('parrot').items():
    print(key, value)
```

```python
def invert_dict(d):
    inv = dict()
    for key in d:
        ## key&val: the key and value of dictionary d
        val = d[key]
        if val not in inv:
            inv[val] = [key]
        else:
            inv[val].append(key)
    return inv


hist = histogram('parrot')
print(hist)
inverted = invert_dict(hist)
print(inverted)
```

## Set

```python
cset = {11, 11, 22}
print(cset)
cset.add(33)
print(cset)
print(id(cset))

# redefine 
cset = {11, 11, 22}
print(id(cset))
print(cset[0]) # set does not support indexing
```

```python
aset = {11, 22,33}
bset = {12, 23,33}
print(aset & bset) # intersection
print(aset | bset) # union
print(aset - bset) # difference
print(aset ^ bset) # symmetric difference
```

### Datetime

```python
import datetime
d1 = datetime.datetime(2026,2,21)
print(d1)
d2 = datetime.datetime(2026, 2, 21, 8, 5, 20)
print(d2.date())
print(d2.time())
print(d2.date().weekday())
d3 = datetime.datetime.today()
diff = d3 - d1
print(diff)

## time shifting
diff2 = datetime.timedelta(days = 2, seconds = 5)
print(d1 + diff2)

# datetime to string
print(d2.strftime('%Y-%m-%d'))
print(d2.strftime('%B %d, %Y'))
print(d2.strftime('%Y-%m-%d %I:%M:%S %p, %A'))
```
