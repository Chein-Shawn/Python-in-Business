## Histogram 直方圖


```python
import csv, datetime

def findSubTimes(fileName):
  fh = open(fileName, "r")
  csvFile = csv.DictReader(fh)

  subTimes = [] # to store submission times
  for row in csvFile:
    # datetime object to time object
    dt = datetime.datetime.strptime(row["SubmissionTime"], "%H:%M:%S").time()
    # from the start of the submission
    sub = (dt.hour - 9) * 3600 + (dt.minute - 20) * 60 + dt.second
    subTimes.append(sub)

  fh.close()
  return subTimes


  
import matplotlib.pyplot as py

subTimes = findSubTimes("midterm2.csv")
# print(subTimes) # just testing
# py.hist 有回傳值
n, bins, patches = py.hist(subTimes, bins = range(0, 12000, 1000))
# py.hist(subTimes, bins = range(0, 12000, 1000), facecolor = "white", edgecolor = "black")
# py.hist(subTimes, bins = 7, color = "gray", edgecolor = "black")
# py.xlim(0, 11000)
# py.ylim(0, 120)
py.xlabel("Submission time")
py.ylabel("Frequency")
py.title("Histogram of submission times")
py.show()

print(n) # the frequencies
print(bins) # the class endpoints
```

## Barchart 長條圖

```python
import csv, datetime

def findSubCnt(filePath):
  fh = open(filePath, "r")
  csvFile = csv.DictReader(fh)

  subCntDict = dict()

  for row in csvFile:
    sid = int(row["StudentID"])
    if sid in subCntDict:
      subCntDict[sid] += 1
    else:
      subCntDict[sid] = 1
    
  fh.close()

  return subCntDict


  
import matplotlib.pyplot as py

subCnt = findSubCnt("midterm2.csv")

# x-axis
seq = range(0, len(subCnt))
width = 0.35
# py.bar(seq, subCnt.values(), width)
# py.bar(seq, sorted(list(subCnt.values())), width)
py.bar(seq, sorted(subCnt.values()), width)

py.show()
```

## Piechart

```python
import csv, datetime

def findProp(filePath):
  fh = open(filePath, "r")
  csvFile = csv.DictReader(fh)

  propDict = dict()

  for row in csvFile:
    result = row["Status"]
    if result in propDict:
      propDict[result] += 1
    else:
      propDict[result] = 1
    
  fh.close()

  return propDict

  
import matplotlib.pyplot as py

pf = findProp("midterm2.csv")

freqs = list(pf.values())
results = list(pf.keys())

#py.pie(freqs, labels = results, autopct = "%1.1f%%")
py.pie(freqs, labels = results, autopct = "%1.2f%%")
py.show()
```
